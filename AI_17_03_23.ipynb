```python
#hoi quy bac 1
import pandas as pd
url ='mtcars.csv'
data = pd.read_csv(url)
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation = data.corr() 
print(correlation)
#gia tri tuyet doi cua cai nao lon nhat thi se phu thuoc vao do
#gia tri am ti le nghich, gia tri duong ti le thuan
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(correlation,annot = True) #bieu dien correlation duoi dang mau sac
```

```python
data.plot(kind='scatter', x = 'wt', y ='mpg', figsize = (9,9), color = 'blue')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X= pd.DataFrame(data['wt']), y=data['mpg'])
print('Gia tri Intercept la : ', model.intercept_)
print('Gia tri Coeffinet la: ', model.coef_)
```

```python
model.score(X= pd.DataFrame(data['wt']), y=data['mpg'])
#score: do chinh xac
X= pd.DataFrame(data['wt'])
```

```python
data.plot(kind='scatter', x = 'wt', y ='mpg', figsize = (9,9), color = 'blue')
y_pred = model.predict(X)
plt.plot(data['wt'],y_pred, color = 'black')
```

```python
#hoi quy bac 2
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'],data['wt']**2]).T #transpose, chuyen vi
y=data['mpg']
model.fit(X,y = data['mpg'])
print('He so Intercept la :', model.intercept_)
print('He so Coeffient la :', model.coef_)
print('Do chinh xac cua mo hinh :',model.score(X, y))
#y = -13,38x + 1,17x^2 + 49,93
# Accuracy = 81%
```

```python
import numpy as np
range = np.arange(1.5, 5.5,0.1)
X_range = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X_range)
data.plot(kind= 'scatter', x = 'wt', y= 'mpg', figsize= (9,9), color = 'black', xlim = (0,7))
plt.plot(range, y_pred, color = 'blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['wt'], data['wt']**2, data['wt']**3]).T
model.fit(X, y = data['mpg'])
print('H s Intercept:', model.intercept_)
print('H s Coeffient:', model.coef_)
print(' chnh xc ca m hnh:', model.score(X, y = data['mpg']))
### y = 0.046(x^3) + 0.69(x^2) - 11.82x + 48.4
### Accurary = 82%
```
