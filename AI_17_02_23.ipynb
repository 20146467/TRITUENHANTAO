```python
#load library
import pandas as pd 
#creta library
df = pd.DataFrame() 
# add columns
df['Name'] = ['Lemon', 'Apple', 'Banana']
df['Price'] = [5, 8 ,7]
df['License'] = [False, True, True] 
#show data
df
```

```python
#số lượng quan sát not-null của mỗi trường trong dataframe
df.info()
```

```python
#giá trị trung bình, độ lệch chuẩn, min, max và tỷ lệ phần trăm của các cột
df.describe()
```
